**DETECT SELECTION IN SEVERAL POPULATIONS**

Manhattan plot is a common tool used in genetic association studies to identify genomic regions associated with a particular 
trait or phenotype. In this case, it helps to identify regions of the genome where mutations are under significant selection
pressure, as indicated by lower log10 p-values. We are focusing on *when_mutation_has_freq2* since when a mutation reaches a
frequency of 2, it indicates that it has persisted in the population for at least two copies. At this point, it is less likely
to be a result of random drift (neutral evolution) and more likely to be under the influence of natural selection.
