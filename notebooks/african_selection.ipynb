**MANHATTAN PLOTS TO DETECT SELECTION IN AFRICAN POPULATIONS**

Manhattan plot is a common tool used in genetic association studies to identify genomic regions associated with a particular 
trait or phenotype. In this case, it helps to identify regions of the genome where mutations are under significant selection
pressure, as indicated by lower log10 p-values. We are focusing on *when_mutation_has_freq2* since when a mutation reaches a
frequency of 2, it indicates that it has persisted in the population for at least two copies. At this point, it is less likely
to be a result of random drift (neutral evolution) and more likely to be under the influence of natural selection. 

```r
library("gridExtra")
library("ggpubr")
library("ggplot2")
```

```r
LWK_plot <- plot(LWK$pos, LWK$when_mutation_has_freq2, pch = 19, col = "lightblue", xlab = "Genomic Position", ylab = "log10(p-value)", main = "Evidence for selection in LWK")
```

```r
# Combine data frames
LWK$Population <- "LWK"
ESN$Population <- "ESN"
GWD$Population <- "GWD"
MSL$Population <- "MSL"
YRI$Population <- "YRI"

combined_data <- rbind(LWK, ESN, GWD, MSL, YRI)

plot_combined <- ggplot(data = combined_data, aes(x = pos, y = when_mutation_has_freq2, color = Population)) +
  geom_point(pch = 19) +
  labs(x = "Genomic Position", y = "log10(p-value)", title = "Evidence for selection in African Populations") +
  theme_light() +
  scale_color_manual(values = c("LWK" = "lightblue", "ESN" = "lightpink", "GWD" = "yellow", "MSL" = "orange", "YRI" = "lightgreen")) +
  theme(
    axis.title = element_text(size = 14),  # Tamaño del texto de los ejes
    axis.text = element_text(size = 12),   # Tamaño del texto de los valores de los ejes
    plot.title = element_text(size = 16, hjust = 0.5)  # Tamaño del texto del título y centrado horizontalmente
  )

# Print the combined plot
print(plot_combined)

ggsave("combined_plot.png", plot_combined, width = 10, height = 6)
```
